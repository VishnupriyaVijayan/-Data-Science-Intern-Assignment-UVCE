```python
# Convert date columns to datetime format
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
```

```python
transactions_df.info(),customers_df.info()
```

```python

# Merge datasets for a comprehensive view
transactions_products = transactions_df.merge(products_df, on='ProductID', how='left')
data = transactions_products.merge(customers_df, on='CustomerID', how='left')
data.head(10)
```

```python
# Check if price_x and price_y are the same
if data['Price_x'].equals(data['Price_y']):
    data.drop(columns=['Price_y'], inplace=True)
```

```python
# Extract year, month, day, and weekday (name) from TransactionDate and SignupDate
data['TransactionDate_Year'] = data['TransactionDate'].dt.year
data['TransactionDate_Month'] = data['TransactionDate'].dt.month
data['TransactionDate_Day'] = data['TransactionDate'].dt.day

data['SignupDate_Year'] = data['SignupDate'].dt.year
data['SignupDate_Month'] = data['SignupDate'].dt.month
data['SignupDate_Day'] = data['SignupDate'].dt.day


data['TransactionDate_Weekday'] = data['TransactionDate'].dt.strftime('%A')  
data['SignupDate_Weekday'] = data['SignupDate'].dt.strftime('%A')  
```

```python
data.drop(columns=['TransactionDate','SignupDate'],inplace=True)
```

```python
# 1. What is the distribution of customers across regions?
region_distribution = data['Region'].value_counts()
region_distribution
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Region',data=data, palette='viridis')
plt.title('Customer Distribution by Region')
plt.ylabel('Number of Customers')
plt.xlabel('Region')
plt.xticks(rotation=35)
plt.show()
```

```python
#2 What is the total quantity, total revenue, times purchased, and average price of each product
product_stats = data.groupby('ProductName').agg(Total_Quantity=('Quantity', 'sum'),
                                                Total_Revenue=('TotalValue', 'sum'),
                                                Times_Purchased=('ProductName', 'count'), 
                                                Average_Price=('Price_x', 'mean')).reset_index()


product_stats = product_stats.sort_values(by='Total_Revenue', ascending=False)

product_stats

```

```python


# Set up the figure size
plt.figure(figsize=(16, 12))

# 1. Transaction Trends
# By Year
plt.subplot(4, 2, 1)
sns.countplot(x='TransactionDate_Year', data=data, palette='viridis')
plt.title('Transactions by Year')
plt.ylabel('Count')
plt.xlabel('Year')

# By Month
plt.subplot(4,2, 2)
sns.countplot(x='TransactionDate_Month', data=data, palette='viridis')
plt.title('Transactions by Month')
plt.ylabel('Count')
plt.xlabel('Month')

# By Day
plt.subplot(4,2, 3)
sns.countplot(x='TransactionDate_Day', data=data, palette='viridis')
plt.title('Transactions by Day')
plt.ylabel('Count')
plt.xlabel('Day')

# By Weekday
plt.subplot(4,2, 4)
sns.countplot(x='TransactionDate_Weekday', data=data, palette='viridis',
              order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Transactions by Weekday')
plt.ylabel('Count')
plt.xlabel('Weekday')

# 2. Signup Trends
# By Year
plt.subplot(4,2, 5)
sns.countplot(x='SignupDate_Year', data=data, palette='viridis')
plt.title('Signups by Year')
plt.ylabel('Count')
plt.xlabel('Year')

# By Month
plt.subplot(4,2, 6)
sns.countplot(x='SignupDate_Month', data=data, palette='viridis')
plt.title('Signups by Month')
plt.ylabel('Count')
plt.xlabel('Month')

# By Day
plt.subplot(4,2, 7)
sns.countplot(x='SignupDate_Day', data=data, palette='viridis')
plt.title('Signups by Day')
plt.ylabel('Count')
plt.xlabel('Day')

# By Weekday
plt.subplot(4,2, 8)
sns.countplot(x='SignupDate_Weekday', data=data, palette='viridis',
              order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Signups by Weekday')
plt.ylabel('Count')
plt.xlabel('Weekday')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
#4. Is there any relation between total value and quantity?
plt.figure(figsize=(8, 4))
sns.boxplot(x='Quantity', y='TotalValue', data=data, palette='viridis')
plt.title('Box Plot of Total Sales Value by Quantity')
plt.xlabel('Quantity')
plt.ylabel('Total Sales Value')
plt.show()
```

```python
#5. Is there any correlation between total value and Price?
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Price_x', y='TotalValue', data=data)
plt.title('Relationship between Quantity and Total Sales Value')
plt.xlabel('Quantity')
plt.ylabel('Total Sales Value')
plt.show()
```

```python
#6. What is the distribution of product prices across categories?
plt.figure(figsize=(10, 4))
sns.boxplot(x='Category', y='Price_x', data=data, palette='viridis')
plt.title('Price Distribution by Product Category')
plt.ylabel('Price')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#7. Which product categories generate the most revenue?
plt.figure(figsize=(10, 4))
category_sales = data.groupby('Category')['TotalValue'].sum().reset_index()
sns.barplot(x='Category', y='TotalValue', data=category_sales, palette='viridis')
plt.title('Total Revenue by Product Category')
plt.ylabel('Total Revenue')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#8. What is the distribution of product category by region?
plt.figure(figsize=(10, 5))
sns.countplot(x='Region', hue='Category', data=data, palette='viridis')
plt.title('Product Distribution by Region')
plt.ylabel('Number of Products')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.show()
```

```python
#9. How does the number of transactions vary across different price ranges?
plt.figure(figsize=(8, 4))
sns.histplot(data['Price_x'], kde=True, color='skyblue')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.show()
```

```python
#10 Aggregate total revenue by Region and Category
region_category_revenue = data.groupby(['Region', 'Category'])['TotalValue'].sum().reset_index()

# Plot grouped bar plot
plt.figure(figsize=(12, 5))
sns.barplot(x='Region', y='TotalValue', hue='Category', data=region_category_revenue, palette='viridis')
plt.title('Revenue Generated by Product Categories Across Regions')
plt.ylabel('Total Revenue')
plt.xlabel('Region')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
# 11. Revenue distribution by region
region_revenue = data.groupby('Region')['TotalValue'].sum()
plt.figure(figsize=(8, 4))
sns.barplot(x=region_revenue.index, y=region_revenue.values, palette='viridis')
plt.title('Revenue by Region')
plt.ylabel('Total Revenue (USD)')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.show()
```

```python
# 12. Top products by revenue
top_products = data.groupby('ProductName')['TotalValue'].sum().nlargest(10)
plt.figure(figsize=(10, 3))
top_products.plot(kind='bar', color='teal')
plt.title('Top 10 Products by Revenue')
plt.ylabel('Total Revenue (USD)')
plt.xlabel('Product Name')
plt.xticks(rotation=80)
plt.show()
```
